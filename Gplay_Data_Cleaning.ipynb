```python
# There is a function of value_counts which shows how many times a value is repeated, so we will use it to see the values repetitions in the dataframe
df['Size'].value_counts()
```

```python
''' 
In the above column, we can see that it has size in it having collection of numbers and alphabets also it has sizes in format of
M, K and Varies with device, it is size like 19MB, etc.
So, we will remove the M for or K for KB associated with the values, and make it in a single format of MB only but not having
any character datatype but convering to the MB size of the data, a single is of size 1024 KB, , after that we do some data cleaning
if any null values are found.
'''
```

```python
# We will create a User-defined function to convert the sizes into float values and KB values into MB values
def clean_size(x):
    if 'k' in x:
        x = x.replace('k','')     # Replacing k with nothing so that we can convert the value in MB
        return float(x)/1024
    elif 'M' in x:
        x = x.replace('M','')
        return float(x)
    elif 'Varies with device' in x:
        return np.nan
    else:
        return float(x)
```

```python
'''
Now we will apply the replace by applying the above function on the column of size by using map function on the size column and mapping our
User-Defined Function with the column of size
'''
```

```python
df['Size'] = df['Size'].map(clean_size)
```

```python
df.info()
```

```python
'''
We can see that the datatype of column Size has now changed from object to float32. Now, after replacing the values we can see 
that null values have been detected in our size column so we replace it by using the same old method
'''
```
